```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Plotting declination as a function of right ascension

RA_degrees, RA_arcminutes, RA_arcseconds = [], [], []       # Initialize empty lists
DEC_degrees, DEC_arcminutes, DEC_arcseconds = [], [], []

lines = open('./assignment1_data/one.txt').readlines()      # Read the file
for line in lines[2:]:
    RA_degrees.append(float(line[25:27]))
    RA_arcminutes.append(float(line[28:30]))
    RA_arcseconds.append(float(line[31:36]))
    DEC_degrees.append(float(line[38:41]))
    DEC_arcminutes.append(float(line[42:44]))
    DEC_arcseconds.append(float(line[45:49]))
```

```python
RA, DEC = [], []    # Initialize empty lists

for i in range(len(RA_arcminutes)):                                             # Convert RA and DEC to degrees
    RA.append(RA_degrees[i] + RA_arcminutes[i]/60 + RA_arcseconds[i]/3600)
    DEC.append(DEC_degrees[i] + DEC_arcminutes[i]/60 + DEC_arcseconds[i]/3600)
```

```python
# Plot declination as a function of right ascension
plt.plot(RA, DEC, 'o', markersize=2)
plt.xlabel('Right Ascension [degrees]')
plt.ylabel('Declination [degrees]')
plt.title('Declination as a function of Right Ascension')
plt.grid(True)
plt.savefig('./assignment1_data/one.png', dpi=100, bbox_inches='tight')
plt.show()
```

```python
distance, half_light_radius = [], []    # Initialize empty lists

with open('./assignment1_data/one.txt') as f:   # Read and process the first file
    lines = f.readlines()
    distance = [float(line[75:80]) for line in lines[2:]]

with open('./assignment1_data/three.txt') as f: # Read and process the third file
    lines = f.readlines()
    for i, line in enumerate(lines[2:]):
        radius_str = line[66:71].strip()  # Strip leading/trailing whitespace
        if radius_str:
            half_light_radius.append(float(radius_str))
        else:
            distance.pop(i)     # Skip invalid data by not appending to half_light_radius and distance
```

```python
# Plot half-light radius as a function of distance
plt.plot(distance, half_light_radius, 'o', markersize=2)
plt.xlabel('Distance from Galactic Center [kpc]')
plt.ylabel('Half-light radius [arcminutes]')
plt.title('Half-light radius as a function of distance from Galactic Center')
plt.grid(True)
plt.xscale('log')
plt.savefig('./assignment1_data/two.png', dpi=100, bbox_inches='tight')
plt.show()
```
